```python
import pandas as pd
import numpy as np
```

```python
train_file='loan_data_train.csv'
test_file='loan_data_test.csv'
```

```python
ld_train=pd.read_csv(train_file)
ld_test=pd.read_csv(test_file)
```

```python
ld_test.head()
```

```python
ld_train.head()
```

```python
ld_train.shape
```

```python
ld_test.shape
```

```python
#note: cant concatinate between 15 columns and 14 columns both should be same
```

```python
ld_test['Interest.Rate']=np.nan
```

```python
ld_train['data'] = 'train'
ld_test['data'] = 'test'
```

```python
ld_train.shape #now a new column is created as Data with train as values
```

```python
ld_test.shape #now a new column is created as Data with test as values and 
            #another column is added as interest rate  with values as np.nan
```

```python
ld_test=ld_test[ld_train.columns] #checking ld_test columns with ld_train column whether they both are matching
#because only when the column name is matching we can concat both .
```

```python
ld_all=pd.concat([ld_train,ld_test],axis=0)
```

```python
ld_all.head()
```

```python
ld_all.dtypes
```

```python
#id,amount funded by investors :drop
#Interest rate&debt to income ratio: remove % 
#FICO range: split into 2 variable and take the average
#employment length : convert to number
```

```python
ld_all.drop(['ID','Amount.Funded.By.Investors'],axis=1,inplace=True)
```

```python
for col in['Interest.Rate','Debt.To.Income.Ratio']:
    ld_all[col]=ld_all[col].str.replace('%','')
```

```python
for col in ['Amount.Requested','Interest.Rate','Debt.To.Income.Ratio','Open.CREDIT.Lines','Revolving.CREDIT.Balance']:
    ld_all[col]=pd.to_numeric(ld_all[col],errors='coerce') #all the mentioned values are given as string should chnage as integer
```

```python
#'errors=coerce' is because if character is there v cannot convert to numeric python will stop hence we are giving an exemption stating the python to proceed even if any character are there
```

```python
ld_all.dtypes #all number that are reflecting as objects previously are changed to float
```

```python
ld_all.head()
```

```python
k=ld_all['FICO.Range'].str.split("-",expand=True).astype(float)#expand splits and creates 2 variables(K0,k1)
ld_all['fico']=0.5*(k[0]+k[1])
del ld_all['FICO.Range'] #deleting this becoz v have created a new fico without range above.
```

```python
ld_all.head()
```

```python
ld_all['Employment.Length'].value_counts() #problem here is +,<,years,year hence we are replacing all this so that the value is numeric
```

```python
ld_all['Employment.Length']=ld_all['Employment.Length'].str.replace('years','')
ld_all['Employment.Length']=ld_all['Employment.Length'].str.replace('year','')
```

```python
ld_all['Employment.Length']=np.where(ld_all['Employment.Length'].str[:2]=='10',10,ld_all['Employment.Length'])
```

```python
ld_all['Employment.Length']=np.where(ld_all['Employment.Length'].str[:0]=='<',0,ld_all['Employment.Length'])
# removing < and replacing it with zero
```

```python
ld_all['Employment.Length']=pd.to_numeric(ld_all['Employment.Length'],errors='coerce')
```

```python
cat_cols=ld_all.select_dtypes(['object']).columns #checking all categorical columns in dataset
```

```python
cat_cols
```

```python
cat_cols=cat_cols[:-1] #removing 'Data' from cat cols since v only created data column.
```

```python
cat_cols
```

```python
ld_all[col]
```

```python
for col in cat_cols:
    freqs=ld_all[col].value_counts()
    k=freqs.index[freqs>20][:-1]
    for cat in k:
        name=col+'_'+cat
        ld_all[name]=(ld_all[col]==cat).astype(int)
    del ld_all[col]
```

```python
ld_all
```

```python
ld_all[name]
```

```python
ld_all.isnull().sum()
```

```python
for col in ld_all.columns:
    if (col not in ['Interest.Rate','data'])& (ld_all[col].isnull().sum()>0):
        ld_all.loc[ld_all[col].isnull(),col]=ld_all.loc[ld_all['data']=='train',col].mean()
```

```python
ld_all.isnull().sum()
```

```python
ld_train=ld_all[ld_all['data']=='train']
del ld_train['data']
ld_test=ld_all[ld_all['data']=='test']
ld_test.drop(['Interest.Rate','data'],axis=1,inplace=True)
```

```python
del ld_all
```

```python
from sklearn.model_selection import train_test_split
```

```python
ld_train1,ld_train2=train_test_split(ld_train,test_size=0.2,random_state=2) 
#since test data doesnt have interest rate column we dont know whether the model is working properly or not
#hence we spilt the train data into 2 parts. train 1 for building model , train 2 for evaluation
#if train1 works properly then it mean it is good
```

```python
x_train1=ld_train1.drop('Interest.Rate',axis=1)
y_train1=ld_train1['Interest.Rate']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train1,y_train1)
```

```python
lm.intercept_
```

```python
list(zip(x_train1.columns,lm.coef_))
```

```python
x_train2=ld_train2.drop('Interest.Rate',axis=1)
```

```python
predicted_ir=lm.predict(x_train2)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(ld_train2['Interest.Rate'],predicted_ir)
```

# RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lambdas=np.linspace(1,100,100)
```

```python
params={'alpha':lambdas}
```

```python
model=Ridge(fit_intercept=True)
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
#grid_search will try all the combination and will find which is best penalty&best estimator
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.cv_results_
```

```python
ridge_model=grid_search.best_estimator_
```

```python
ridge_model.fit(x_train,y_train)
```

```python
list(zip(x_train1.columns,ridge_model.coef_))
```

```python
test_pred=grid_search.predict(ld_test)
```

```python
pd.DataFrame(test_pred).to_csv("mysubmission.csv",index=False)
```

# LASSO REGRESSION

```python
lambdas=np.linspace(1,10,100)
model=Lasso(fit_intercept=True)
params={'alpha':lambdas}
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```
